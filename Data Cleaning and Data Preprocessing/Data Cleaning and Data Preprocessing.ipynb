# Data Cleaing and Preprocessing

Rename the names from the column Name

```python
df["name"] = df["name"].str.split(" ").str.slice(0, 3).str.join(" ")
df.name
```

```python
df["name"].unique()
```

```python
df.info()
```

```python
df["company"].unique()
```

```python
df.info()
```

Removing String Values and Convert the Year Column to Integer

```python
df = df[df["year"].str.isnumeric()]
df.year
```

```python
df.year =  df["year"].astype(int)
df.year
```

```python
df["year"].unique()
```

```python
df.info()
```

Removing String and Commas and Convert the Price Column to Integer

```python
df = df[df["Price"] != "Ask For Price"]
df.Price
```

```python
df.Price = df.Price[:].str.replace(",","")
df.Price
```

```python
df.Price = df.Price.astype(int)
df.Price
```

```python
df["Price"].unique()
```

```python
df.info()
```

Removing Commas, kms and Nan Value and Convert the Kms_Driven Column to Integer

```python
df.kms_driven = df.kms_driven[:].str.replace(',','')
df.kms_driven
```

```python
df.kms_driven = df.kms_driven[:].str.replace('kms','')
df.kms_driven
```

```python
df.kms_driven = pd.to_numeric(df["kms_driven"], errors = "coerce")
df.kms_driven
```

```python
df = df[~df["kms_driven"].isna()]
df.kms_driven
```

```python
df.kms_driven = df['kms_driven'].astype(int)
df.kms_driven
```

```python
df["kms_driven"].unique()
```

```python
df.info()
```

Removing Null Value From The Fuel-Type Column

```python
df = df[~df["fuel_type"].isna()]
df.fuel_type
```

```python
df["fuel_type"].unique()
```

```python
df.info()
```
